Calculating summary statistics 

```python
import pandas as pd

# Load and preprocess data
data = pd.read_csv("final_thesis_data.csv")

# List the continuous numeric variables and categorical variables
numeric_variables = [
    "asy_applications", 
    "illegal_border_crossings", 
    "push_factor_index", 
    "push_factor_index_high_level",
    "deaths_civilians", 
    "regime_duration", 
    "gdp_per_capita_current_usd", 
    "gdp_per_capita_growth", 
    "regime_end_type",
    "state_fiscal_source_revenue",
    "state_authority_over_territory",
    "political_polarisation",
    "political_violence",
    "domestic_autonomy",
    "rule_of_law",
]

categorical_variables = ["sex", "age_group"]

# Check for missing values in numeric variables
missing_values_numeric = data[numeric_variables].isna().sum()

# Calculate descriptive statistics for each numeric variable
descriptive_stats = data[numeric_variables].describe().T

# Add the missing values information to the descriptive statistics
descriptive_stats['Missing Values'] = missing_values_numeric

# Check for missing values in categorical variables
missing_values_categorical = data[categorical_variables].isna().sum()

# Create a DataFrame for categorical variables with missing values
categorical_missing_df = pd.DataFrame(missing_values_categorical, columns=['Missing Values'])

# Combine numeric and categorical missing values
combined_stats = pd.concat([descriptive_stats, categorical_missing_df])

# Save the descriptive statistics and missing values table to a CSV file
combined_stats.to_csv("descriptive_statistics_with_missing_values.csv")

# Display the table to the user
print("Descriptive statistics and missing values for variables:")
print(combined_stats)
```
